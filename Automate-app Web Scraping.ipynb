```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support.ui import Select 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from datetime_selenium import send_datetime
from datetime import datetime
```

```python
driver = webdriver.Chrome()
website = 'https://www.automate-app.com/en/insurance/form-new'
driver.get(website)
driver.maximize_window()
wait = WebDriverWait(driver, 30)
action = webdriver.common.action_chains.ActionChains(driver)
datetime_ = datetime.utcnow()



def initial_prompts():
    wait.until(EC.element_to_be_clickable(('xpath', '//*[@id="__next"]/div/div[1]/div/div/div/div[1]/div[3]/div/div')))
    driver.find_element('xpath', '//*[@id="__next"]/div/div[1]/div/div/div/div[1]/div[3]/div/div').click()
    driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div[1]/div/div[2]').click()
    time.sleep(1)
    driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div[2]/div/div[1]').click()
    time.sleep(1)
    driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div[3]/div/div[2]').click()
    time.sleep(1)
    driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[2]/div[3]').click()
    
    
def year_2018():
    year = driver.find_element('xpath', '//*[@id="carYear"]')
    year.send_keys('2018')

    
def select_car_make(x):
    car_make = driver.find_element('xpath', '//*[@id="carMake"]')
    car_make.send_keys(f'{x}')
    driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[2]/div/div/div').click()

def final_prompts():
    engine = driver.find_element('xpath', '//*[@id="engine_capacity"]')
    engine.send_keys('1300')
    time.sleep(1)
    value = driver.find_element('xpath', '//*[@id="market_value"]')
    value.send_keys('50000')
    value.send_keys(Keys.ENTER)
    time.sleep(1)
    ncd = driver.find_element('xpath', '//*[@id="ncd"]')
    ncd.send_keys('60')
    time.sleep(1)
    choose_60 = driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[6]/div/div/div/p')
    ActionChains(driver).move_to_element(choose_60).click(choose_60).perform()
    time.sleep(1)
    effective_date = driver.find_element('xpath', '//*[@id="headlessui-popover-button-6"]/button/time[@datetime="2023-07-22"]').click()
    time.sleep(1)
    fully_paid = driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[8]/div[2]/div/div[1]/p').click()
    
def select_car_model():
    dropdown_list = wait.until(EC.presence_of_all_elements_located(('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[3]/div/div[2]/div')))
    for i in range(len(dropdown_list)):
        driver.find_element('xpath', f'//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[3]/div/div[2]/div[{i+1}]').click()
        if i < 1:
            final_prompts()
        get_quote = driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[2]/div[3]/p').click()
        try:
            wait.until(EC.invisibility_of_element_located(('xpath', '//*[@id="loader"]')))
            time.sleep(1)
            wait.until(EC.presence_of_element_located(('xpath', '//div[@class="lg:block"]/div[@class="mb-5"]//p[@class="hidden lg:block font-medium text-lg lg:text-2-1/3xl"]')))
            time.sleep(1)
            wait.until(EC.element_to_be_clickable(('xpath', '//div[@class="lg:block"]/div[@class="mb-5"]//p[@class="hidden lg:block font-medium text-lg lg:text-2-1/3xl"]')))
            prices = driver.find_elements('xpath', '//div[@class="lg:block"]/div[@class="mb-5"]//p[@class="hidden lg:block font-medium text-lg lg:text-2-1/3xl"]')
            print(prices)
            for price in prices:
                print(price.text)
            go_back = driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[2]/div[2]/div').click()
            time.sleep(2)
            wait.until(EC.element_to_be_clickable(('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[3]/div/*[name()="svg"][@class="cursor-pointer"]'))).click()
        except:
            go_back = driver.find_element('xpath', '//*[@id="__next"]/div/div/div/main/div[2]/div[2]/div').click()
            time.sleep(2)
            wait.until(EC.element_to_be_clickable(('xpath', '//*[@id="__next"]/div/div/div/main/div[3]/div[2]/div/div[1]/div[1]/div/div[3]/div/*[name()="svg"][@class="cursor-pointer"]'))).click()

        
def retrieve_insurance_prices(x):
    initial_prompts()
    time.sleep(1)
    year_2018()
    time.sleep(1)
    select_car_make(f"{x}")
    time.sleep(1)
    select_car_model()

retrieve_insurance_prices("Lexus")
```
