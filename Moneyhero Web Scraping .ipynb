```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support.ui import Select 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
```

```python
driver = webdriver.Chrome()
website = 'https://insurance.moneyhero.com.hk/en/car/quote'
driver.get(website)
driver.maximize_window()
wait = WebDriverWait(driver, 30)
action = webdriver.common.action_chains.ActionChains(driver)
wait.until(EC.text_to_be_present_in_element(('xpath', '//*[@id="basic_form_section"]/div/div[2]/div'), 'My car is'))


def select_year_2018():
    driver.execute_script("window.scrollTo(0, 150)")
    year_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[2]/bh-selectinput').click()
    time.sleep(3)
    year_2018 = driver.find_element('xpath', '//*[@id="mat-menu-panel-1"]/div/button[6]').click()

def select_car_make(x):
    car_make_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[2]/bh-selectinput[2]/button').click()
    time.sleep(3)
    driver.find_element('xpath', f'//*[@id="mat-menu-panel-2"]/div/button[{x}]').click()

def select_car_model(x):
    car_model_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[2]/bh-selectinput[3]')
    car_model_button.click()
    car_model_list = driver.find_elements('xpath', '//*[@id="mat-menu-panel-3"]/div/button')
    action.move_by_offset(0, 150).click().perform()
    for i in range(len(car_model_list)):
        select_year_2018()
        select_car_make(x)
        driver.execute_script("window.scrollTo(0, 150)")
        time.sleep(3)
        car_model_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[2]/bh-selectinput[3]').click()
        print("Model", i)
        time.sleep(3)
        try:
            driver.find_element('xpath', f'//*[@id="mat-menu-panel-3"]/div/button[{i+1}]').click()
            print(driver.find_element('xpath', f'//*[@id="mat-menu-panel-3"]/div/button[{i+1}]').text)
            time.sleep(3)
            cylinder = driver.find_element('xpath', '//*[@id="mat-input-0"]')
            sum_insured = driver.find_element('xpath', '//*[@id="mat-input-1"]')
            driving_years_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[3]/bh-selectinput/button')
            years_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[3]/span/bh-selectinput/button')
            ncd = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[4]/bh-selectinput/button')
            cylinder.clear()
            cylinder.send_keys('1300')
            time.sleep(1)
            sum_insured.clear()
            sum_insured.send_keys('50000')
            time.sleep(1)
            driving_years_button.click()
            time.sleep(1)
            two_years_experience = driver.find_element('xpath', '//*[@id="mat-menu-panel-4"]/div/button[1]').click()
            time.sleep(1)
            years_button.click()
            time.sleep(1)
            twenty_six_years = driver.find_element('xpath', '//*[@id="mat-menu-panel-5"]/div/button[2]').click()
            time.sleep(1)
            ncd.click()
            time.sleep(1)
            ncd_60 = driver.find_element('xpath', '//*[@id="mat-menu-panel-6"]/div/button[6]').click()
            time.sleep(1)
            next_button = driver.find_element('xpath', '//*[@id="basic_form_section"]/div/div[6]/bh-basicbutton/button').click()
            try:
                wait.until(EC.invisibility_of_element_located(('xpath', '//mat-progress-spinner')))
                waiting_elements = wait.until(EC.presence_of_all_elements_located(('xpath', '/html/body/app-root/bh-list_insurance/div/div[2]/div[2]/bh-products_list/bh-product_card/div/div[2]/div[1]/div[3]/div[4]/bh-premium/div/div/span')))
                prices = driver.find_elements('xpath', '/html/body/app-root/bh-list_insurance/div/div[2]/div[2]/bh-products_list/bh-product_card/div/div[2]/div[1]/div[3]/div[4]/bh-premium/div/div/span')
                insurances = driver.find_elements('xpath', '//h2[@class="product-card-title"]')
                for price in prices:
                    print(price.text)
                for insurance in insurances:
                    print(insurance.text)
                driver.get(website)
                wait.until(EC.text_to_be_present_in_element(('xpath', '//*[@id="basic_form_section"]/div/div[2]/div'), 'My car is'))
            except: 
                driver.get(website)
                wait.until(EC.text_to_be_present_in_element(('xpath', '//*[@id="basic_form_section"]/div/div[2]/div'), 'My car is'))
        except: pass
    
def retrieve_prices(x):
    select_year_2018()
    time.sleep(3)
    select_car_make(x)
    time.sleep(3)
    select_car_model(x)

retrieve_prices(3)
```
