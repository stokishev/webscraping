```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support.ui import Select 
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from datetime_selenium import send_datetime
from datetime import datetime
import pandas as pd
```

```python
driver = webdriver.Chrome()
base_url = 'https://propwiser.com.hk/tc/Building?page='
driver.maximize_window()
wait = WebDriverWait(driver, 10)

building_total = []
building_df = pd.DataFrame([],columns= ['name','name_eng','area'])
        
        
def scrape_page():
    looping_index = driver.find_elements('xpath', '//*[@id="maincontent"]/div[2]/div[4]/ul/li/a/div[2]')
    for i in range(len(looping_index)):
        building_name = driver.find_element('xpath', f'//*[@id="maincontent"]/div[2]/div[4]/ul/li[{i+1}]/a/div[2]').text
        print(building_name)
        building_name_english = driver.find_element('xpath', f'//*[@id="maincontent"]/div[2]/div[4]/ul/li[{i+1}]/a/div[3]').text
        print(building_name_english)
        try:
            building_area = driver.find_element('xpath', f'//*[@id="maincontent"]/div[2]/div[4]/ul/li[{i+1}]/a/div[4]').text
            print(building_area)
            building_total.append([building_name, building_name_english, building_area])
        except: 
            building_total.append([building_name, building_name_english])
    return building_total


for index in range(1,39):
    print('Trial', index, '\n')
    url = base_url + str(index)
    driver.get(url)
    wait.until(EC.presence_of_element_located(('xpath', '//*[@id="maincontent"]/div[2]/div[4]/ul/li/a/div[2]')))
    time.sleep(1)
    scrape_page()
```

```python
df = building_df.append(building_total)
```

```python
df.drop('name_eng', inplace=True, axis=1)
df.drop('area', inplace=True, axis=1)
```

```python
df.rename(columns = {0:'name'}, inplace = True)
df.rename(columns = {1:'name_eng'}, inplace = True)
df.rename(columns = {2:'area'}, inplace = True)
```
